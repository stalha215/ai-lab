```python
import heapq

def uniform_cost_search(graph, start, goal):
    # Priority Queue stores (cumulative_cost, current_node, path_taken)
    priority_queue = [(0, start, [start])]
    # Set to keep track of visited nodes to avoid cycles
    visited = set()

    print(f"Starting UCS from {start} to {goal}...\n")

    while priority_queue:
        # Pop the node with the lowest cost
        (cost, current_node, path) = heapq.heappop(priority_queue)

        if current_node in visited:
            continue

        print(f"Visiting: {current_node} | Cumulative Cost: {cost}")

        # Check if we reached the goal
        if current_node == goal:
            return path, cost

        visited.add(current_node)

        # Explore neighbors
        for neighbor, weight in graph.get(current_node, {}).items():
            if neighbor not in visited:
                total_cost = cost + weight
                heapq.heappush(priority_queue, (total_cost, neighbor, path + [neighbor]))

    return None, float('inf')

# --- Define your Graph (Adjacency List with Weights) ---
# Format: 'Node': {'Neighbor': Weight}
graph = {
    'Islamabad': {'Rawalpindi': 20, 'Lahore': 300, 'Peshawar': 180},
    'Rawalpindi': {'Quetta': 900, 'Islamabad': 20},
    'Peshawar': {'Quetta': 800, 'Islamabad': 180},
    'Lahore': {'Multan': 350, 'Quetta': 1000},
    'Multan': {'Karachi': 900, 'Lahore': 350, 'Quetta': 600},
    'Quetta': {'Karachi': 600, 'Multan': 600},
    'Karachi': {'Multan': 900, 'Quetta': 600}
}

# --- Execute the Search ---
start_node = 'Islamabad'
destination = 'Karachi'

best_path, total_min_cost = uniform_cost_search(graph, start_node, destination)

if best_path:
    print(f"\n✅ Shortest Path Found: {' -> '.join(best_path)}")
    print(f"💰 Total Minimum Cost: {total_min_cost}")
else:
    print("\n❌ No path found.")
```
