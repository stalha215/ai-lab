```python
import random
import time
import sys
import collections

# Increase recursion depth for deep DFS
sys.setrecursionlimit(2000000)

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, key):
        if self.root is None:
            self.root = Node(key)
        else:
            self._insert_recursive(self.root, key)

    def _insert_recursive(self, root, key):
        if key < root.val:
            if root.left is None:
                root.left = Node(key)
            else:
                self._insert_recursive(root.left, key)
        else:
            if root.right is None:
                root.right = Node(key)
            else:
                self._insert_recursive(root.right, key)

def bfs(root, goal):
    if not root: return False
    queue = collections.deque([root])
    while queue:
        node = queue.popleft()
        if node.val == goal:
            return True
        if node.left: queue.append(node.left)
        if node.right: queue.append(node.right)
    return False

def dfs_recursive(root, goal):
    if not root: return False
    if root.val == goal: return True
    return dfs_recursive(root.left, goal) or dfs_recursive(root.right, goal)

# Main Execution Loop
input_sizes = [1000, 40000, 80000, 200000, 1000000]
max_val_range = 10000000

for size in input_sizes:
    print(f"\n--- Set Size: {size} ---")

    # Generate and build tree
    data_set = random.sample(range(1, max_val_range), size)
    goal_node = data_set[-1] # Set the last inserted node as the goal

    tree = BinarySearchTree()
    for num in data_set:
        tree.insert(num)

    # 1. Measure BFS Time
    start_bfs = time.time()
    bfs_found = bfs(tree.root, goal_node)
    end_bfs = time.time()
    bfs_time = end_bfs - start_bfs

    # 2. Measure DFS Time
    start_dfs = time.time()
    dfs_found = dfs_recursive(tree.root, goal_node)
    end_dfs = time.time()
    dfs_time = end_dfs - start_dfs

    print(f"BFS Time: {bfs_time:.6f} seconds")
    print(f"DFS Time: {dfs_time:.6f} seconds")
```

```python
import random
import time
import sys
import collections

# Increase recursion depth for deep DFS on large trees
sys.setrecursionlimit(2000000)

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, key):
        if self.root is None:
            self.root = Node(key)
        else:
            self._insert_recursive(self.root, key)

    def _insert_recursive(self, root, key):
        if key < root.val:
            if root.left is None:
                root.left = Node(key)
            else:
                self._insert_recursive(root.left, key)
        else:
            if root.right is None:
                root.right = Node(key)
            else:
                self._insert_recursive(root.right, key)

def bfs_search(root, goal):
    if not root: return False
    queue = collections.deque([root])
    while queue:
        node = queue.popleft()
        if node.val == goal:
            return True
        if node.left: queue.append(node.left)
        if node.right: queue.append(node.right)
    return False

def dfs_search(root, goal):
    if not root: return False
    if root.val == goal: return True
    # Standard DFS (Pre-order style)
    if dfs_search(root.left, goal): return True
    return dfs_search(root.right, goal)

# Main Experiment
input_sizes = [1000, 40000, 80000, 200000, 1000000]
max_range = 10000000

print(f"{'Size':<10} | {'BFS Time (s)':<15} | {'DFS Time (s)':<15}")
print("-" * 45)

for size in input_sizes:
    # 1. Generate unique random numbers
    data_list = random.sample(range(1, max_range), size)

    # 2. Define the Goal based on the required index: [total_len - 220]
    # We use data_list because the tree nodes aren't indexed like an array
    goal_val = data_list[size - 220]

    # 3. Build the Tree
    tree = BinarySearchTree()
    for x in data_list:
        tree.insert(x)

    # 4. Measure BFS Time
    start = time.time()
    bfs_search(tree.root, goal_val)
    bfs_duration = time.time() - start

    # 5. Measure DFS Time
    start = time.time()
    dfs_search(tree.root, goal_val)
    dfs_duration = time.time() - start

    print(f"{size:<10} | {bfs_duration:<15.6f} | {dfs_duration:<15.6f}")
```

```python
import random
import time
import sys
import collections
import matplotlib.pyplot as plt
import numpy as np

# Increase recursion limit just in case
sys.setrecursionlimit(2000000)

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

# Iterative insertion to handle 1,000,000 nodes without stack overflow
def insert_iterative(root, key):
    if root is None:
        return Node(key)
    curr = root
    while True:
        if key < curr.val:
            if curr.left is None:
                curr.left = Node(key)
                break
            curr = curr.left
        else:
            if curr.right is None:
                curr.right = Node(key)
                break
            curr = curr.right
    return root

# Iterative BFS
def bfs_search(root, goal):
    if not root: return False
    queue = collections.deque([root])
    while queue:
        node = queue.popleft()
        if node.val == goal:
            return True
        if node.left: queue.append(node.left)
        if node.right: queue.append(node.right)
    return False

# Iterative DFS (using a manual stack)
def dfs_search_iterative(root, goal):
    if not root: return False
    stack = [root]
    while stack:
        node = stack.pop()
        if node.val == goal:
            return True
        # Push right then left to maintain left-to-right order
        if node.right: stack.append(node.right)
        if node.left: stack.append(node.left)
    return False

# Configuration
input_sizes = [1000, 40000, 80000, 200000, 1000000]
max_range = 10000000
bfs_times = []
dfs_times = []

print(f"{'Size':<10} | {'BFS Time (s)':<15} | {'DFS Time (s)':<15}")
print("-" * 45)

for size in input_sizes:
    # 1. Generate unique random numbers
    data_list = random.sample(range(1, max_range), size)

    # 2. Set Goal as per requirement: list[total_len - 220]
    goal_val = data_list[size - 220]

    # 3. Build the Tree
    root = None
    for x in data_list:
        if root is None:
            root = Node(x)
        else:
            insert_iterative(root, x)

    # 4. Measure BFS Time
    start = time.time()
    bfs_search(root, goal_val)
    bfs_duration = time.time() - start
    bfs_times.append(bfs_duration)

    # 5. Measure DFS Time
    start = time.time()
    dfs_search_iterative(root, goal_val)
    dfs_duration = time.time() - start
    dfs_times.append(dfs_duration)

    print(f"{size:<10} | {bfs_duration:<15.6f} | {dfs_duration:<15.6f}")

# --- Plotting the Bar Chart ---
x = np.arange(len(input_sizes))  # Label locations
width = 0.35  # Width of the bars

fig, ax = plt.subplots(figsize=(12, 7))
rects1 = ax.bar(x - width/2, bfs_times, width, label='BFS', color='skyblue')
rects2 = ax.bar(x + width/2, dfs_times, width, label='DFS', color='salmon')

# Add labels and title
ax.set_ylabel('Time in Seconds')
ax.set_xlabel('Input Data Set Size')
ax.set_title('Comparison of BFS vs DFS Execution Time')
ax.set_xticks(x)
ax.set_xticklabels([f"{s:,}" for s in input_sizes])
ax.legend()

# Optional: Add grid for better readability
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
